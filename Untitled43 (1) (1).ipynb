```python
train_texts, test_texts, train_labels, test_labels = train_test_split(
    df['text'].values,
    df['label'].values,
    test_size=0.2,
    random_state=42
)
```

```python
import torch
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("cardiffnlp/twitter-roberta-base-sentiment")

def tokenize_data(texts):
    return tokenizer(list(texts), padding=True, truncation=True, max_length=128, return_tensors="pt")

train_encodings = tokenize_data(train_texts)
test_encodings = tokenize_data(test_texts)
```

```python
class TwitterDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: val[idx] for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx], dtype=torch.long) # Changed dtype to torch.long
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = TwitterDataset(train_encodings, train_labels)
test_dataset = TwitterDataset(test_encodings, test_labels)
```

```python
import tempfile
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer
import evaluate
import numpy as np
```

```python
metric = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)
```

```python
model = AutoModelForSequenceClassification.from_pretrained(
    "cardiffnlp/twitter-roberta-base-sentiment",
    num_labels=2,
    ignore_mismatched_sizes=True
)
temp_dir = tempfile.mkdtemp()
training_args = TrainingArguments(
    output_dir=temp_dir,
    overwrite_output_dir=True,
    num_train_epochs=3,
    per_device_train_batch_size=16,
    save_strategy="no",
    eval_strategy="epoch",
    learning_rate=2e-5,
    report_to="none"
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    compute_metrics=compute_metrics,
)
```

```python
print(f"Training has started in the temporary directory: {temp_dir}")
trainer.train()
```

```python
save_dir = "/content/sentiment_roberta_model"

trainer.save_model(save_dir)
tokenizer.save_pretrained(save_dir)

print(f"Model saved to {save_dir}")
```

```python
import os
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# Load the fine-tuned model and tokenizer
model_path = "/content/sentiment_roberta_model"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSequenceClassification.from_pretrained(model_path, num_labels=2)

# Ensure model is in evaluation mode
model.eval()

onnx_model_path = "/content/onnx_model/model.onnx"
onnx_model_dir = os.path.dirname(onnx_model_path)
os.makedirs(onnx_model_dir, exist_ok=True)

# Create dummy inputs for ONNX export
dummy_input = tokenizer("This is a test sentence.", return_tensors="pt", truncation=True, padding=True, max_length=128)

# Export the model to ONNX
with torch.no_grad():
    torch.onnx.export(
        model,
        (dummy_input["input_ids"], dummy_input["attention_mask"]),
        onnx_model_path,
        input_names=["input_ids", "attention_mask"],
        output_names=["logits"],
        dynamic_axes={
            "input_ids": {0: "batch_size", 1: "sequence_length"},
            "attention_mask": {0: "batch_size", 1: "sequence_length"},
            "logits": {0: "batch_size"}
        },
        opset_version=13,
        do_constant_folding=True
    )

print(f"Model exported to ONNX at: {onnx_model_path}")
```

```python
import onnxruntime as ort
import torch
import numpy as np
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("/content/sentiment_roberta_model")

session = ort.InferenceSession("/content/onnx_model/model.onnx")

def predict_sentiment_onnx(text):
    inputs = tokenizer(
        text,
        return_tensors="np",
        truncation=True,
        padding=True,
        max_length=128
    )

    ort_inputs = {
        session.get_inputs()[0].name: inputs["input_ids"],
        session.get_inputs()[1].name: inputs["attention_mask"]
    }

    logits = session.run(None, ort_inputs)[0]
    probs = torch.softmax(torch.tensor(logits), dim=-1)

    labels = ["Negative", "Positive"]
    idx = torch.argmax(probs).item()
    confidence = probs[0][idx].item() * 100

    return labels[idx], confidence
```

```python
text = "This project is amazing and very well done!"
sentiment, score = predict_sentiment_onnx(text)

print(sentiment, f"{score:.2f}%")
```

```python
import torch
import torch.nn.functional as F
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"

def predict_sentiment(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=128).to(device)
    with torch.no_grad():
        outputs = model.to(device)(**inputs)
    probs = F.softmax(outputs.logits, dim=-1)

    labels = ['Negative ', 'Positive ']
    label_idx = torch.argmax(probs).item()
    confidence = probs[0][label_idx].item() * 100

    return labels[label_idx], confidence
```

```python
test_list = [
    "This new AI project is absolutely mind-blowing! I've never seen such performance before.",
    "I've been waiting for the update for 5 hours and it still crashes every time. So frustrating!",
    "Oh great, another Monday with a broken coffee machine. Just what I needed."
]

print("="*50)
print(f"Testing on Device: {device}")
print("="*50)

for tweet in test_list:
    sentiment, score = predict_sentiment(tweet)
    print(f"Tweet: {tweet}")
    print(f"Result: {sentiment} ({score:.2f}%)")
    print("-" * 50)
```
